```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

import xgboost as xgb
```

```python
# 1. Load Dataset
# ================================
df = pd.read_csv("comprehensive_mutual_funds_data.csv")
```

```python
# 2. Data Cleaning
# ================================

# Convert object columns (alpha, beta, sharpe, sortino, sd) to numeric
cols_to_convert = ["sortino", "alpha", "sd", "beta", "sharpe"]
for col in cols_to_convert:
    df[col] = pd.to_numeric(df[col], errors="coerce")

# Handle missing values (fill with median)
df.fillna(df.median(numeric_only=True), inplace=True)
```

```python
# 3. Risk Level Mapping (Target Variable)
# ================================
# Risk levels (1-2: Low, 3-4: Medium, 5-6: High)
def map_risk(x):
    if x <= 2:
        return "Low"
    elif x <= 4:
        return "Medium"
    else:
        return "High"

df["risk_category"] = df["risk_level"].apply(map_risk)

# Encode target for ML
le = LabelEncoder()
df["risk_encoded"] = le.fit_transform(df["risk_category"])
```

```python
# 4. Feature Selection
# ================================
features = [
    "expense_ratio", "fund_size_cr", "fund_age_yr",
    "sortino", "alpha", "sd", "beta", "sharpe",
    "returns_1yr", "returns_3yr", "returns_5yr", "rating"
]

X = df[features]
y = df["risk_encoded"]

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# 5. Train-Test Split
# ================================
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)
```

```python
# 6. XGBoost Classification
# ================================
xgb_model = xgb.XGBClassifier(
    objective="multi:softmax",
    num_class=3,
    eval_metric="mlogloss",
    random_state=42
)

xgb_model.fit(X_train, y_train)
y_pred = xgb_model.predict(X_test)

# Evaluation
print("\nðŸ“Š Classification Report:")
print(classification_report(y_test, y_pred, target_names=le.classes_))
print("âœ… Accuracy:", accuracy_score(y_test, y_pred))

# Confusion Matrix
plt.figure(figsize=(6,4))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt="d",
            xticklabels=le.classes_, yticklabels=le.classes_, cmap="Blues")
plt.title("Confusion Matrix - XGBoost Risk Classification")
plt.show()

# Feature Importance
xgb.plot_importance(xgb_model, importance_type="weight")
plt.title("Feature Importance - XGBoost")
plt.show()
```

```python
# 7. Clustering with KMeans + PCA
# ================================
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X_pca)
df["cluster"] = clusters

# Visualize Clusters
plt.figure(figsize=(8,6))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=clusters, palette="Set1")
plt.title("KMeans Clustering of Mutual Funds (PCA Reduced)")
plt.show()
```

```python
# 8. Portfolio Recommendation System
# ================================
def recommend_portfolio(risk_pref, top_n=5):
    """
    Recommend top funds based on investor risk preference
    risk_pref: "Low", "Medium", "High"
    """
    # Filter by risk preference
    subset = df[df["risk_category"] == risk_pref]

    # Rank by returns (3yr) + rating
    subset = subset.copy()
    subset["score"] = subset["returns_3yr"]*0.7 + subset["rating"]*0.3
    top_funds = subset.sort_values("score", ascending=False).head(top_n)

    return top_funds[["scheme_name", "amc_name", "category", "sub_category",
                      "returns_1yr", "returns_3yr", "returns_5yr", "rating", "risk_category"]]
```

```python
# Example usage:
print("\nðŸ’¡ Recommended Portfolio for MEDIUM risk investors:")
print(recommend_portfolio("Low", top_n=2))
```
